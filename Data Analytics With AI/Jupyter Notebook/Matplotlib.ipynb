```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Scatter plot
```

```python
x=np.random.rand(200)*100
y=np.random.rand(200)*100
```

```python
print(y)
```

```python
plt.scatter(x,y,color="red",edgecolor="black",s=70,alpha=0.5)
plt.title("scatter plot example data",fontsize=12,color="blue")
plt.xlabel("x example values")
plt.ylabel("y example values")
plt.grid(True,alpha=0.9,linestyle="--")
plt.show()
```

```python
#line Chart
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
weights = [58.2, 59.0, 60.1, 61.4, 61.8, 62.5, 63.1, 64.0, 64.7, 65.3, 65.8, 66.2, 66.8, 67.1, 70.5]
```

```python

plt.plot(years,weights,color="green",marker=">",linestyle="--",lw=2,markersize=7)
plt.title("weight change over years",fontsize=12,color="blue")
plt.xlabel("years")
plt.ylabel("weights")
plt.grid(True,alpha=0.9,linestyle="--")
plt.show()
```

```python
#Bar Graph

categories = ['Electronics', 'Clothing', 'Groceries', 'Furniture', 'Toys', 'Books', 'Sports']
sales = [120, 85, 150, 95, 60, 40, 70]
plt.bar(categories,sales,color="blue",edgecolor="red",width=0.9)
plt.title("sales of categories",fontsize=12,color="blue")
plt.xlabel("categories")
plt.ylabel("sales amount")
plt.grid(True,alpha=0.4,linestyle="--")


plt.show()

```

```python
#Histogram
x=np.random.rand(200)*100
```

```python
plt.hist(x,bins=15,color="#0D13A6",edgecolor="blue")
plt.title("sales amount",fontsize=12,color="blue")
plt.xlabel("sales")
plt.ylabel("sales amount")
plt.grid(True,alpha=0.4,linestyle="--")
plt.tight_layout()

plt.show()
```

```python
#pie chart
categories = ['Electronics', 'Clothing', 'Groceries', 'Furniture', 'Toys', 'Books', 'Sports']
sales = [120, 85, 150, 95, 60, 40, 70]
plt.pie(sales,labels=categories,autopct="%.1f%%",startangle=90)
plt.title("product sales",fontsize=12,color="blue")
plt.grid(True,alpha=0.4,linestyle="--")
plt.tight_layout()

plt.show()
```

```python
#box plot
```

```python
data=np.random.normal(165,6,200)
```

```python
print(data)
```

```python
plt.boxplot([data],patch_artist=True,boxprops=dict(facecolor="red"),whiskerprops=dict(color="red"),capprops=dict(color="blue"),flierprops=dict(color="red"))
plt.title("average height distribution")
plt.show()
```

```python
stock1 = [120, 85, 150, 95, 60, 40, 70]
stock2 = [100, 35, 120, 75, 80, 80, 20]
stock3 = [150, 65, 110, 25, 60, 20, 10]

plt.plot(stock1,label="company1")
plt.plot(stock2,label="company2")
plt.plot(stock3,label="company3")

plt.legend()
plt.show()
```

```python
Weight=[23,45,12,78,27,54]
people=["A","B","C","D","E","F"]
plt.pie(Weight,labels=None)
plt.legend(labels=people,loc="lower right")
plt.show()
```

```python
ax=plt.axes(projection="3d")
x=np.random.random(100)
y=np.random.random(100)
z=np.random.random(100)
ax.scatter(x,y,z)
ax.set_title("3d data for scatter")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```
